```python
# Example of using isinstance() to validate input
def process(item):
    # Type guard: check if item is an integer
    if isinstance(item, int):
        return item * 2
    # Type guard: check if item is a string
    elif isinstance(item, str):
        return item.upper()

# Test the function
result1 = process(5)
result2 = process("hello")
print(f"process(5) = {result1}")
print(f"process('hello') = {result2}")
```
